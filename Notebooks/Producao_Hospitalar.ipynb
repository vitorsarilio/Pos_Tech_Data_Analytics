```python
dados_jun_2023= dados[["Unidade da Federação","2023/Jun"]]
```

```python
dados_jun_2023 = dados_jun_2023.drop(5)
```

```python
dados_jun_2023.info()
```

```python
dados_jun_2023["2023/Jun"] = dados_jun_2023["2023/Jun"].astype(float)
dados_jun_2023.plot(x="Unidade da Federação",y="2023/Jun")
```

```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

axis = dados_jun_2023.plot(x="Unidade da Federação",y="2023/Jun",kind='bar',figsize=(20,6))
axis.yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.2f}"))

plt.tick_params(axis='x', labelrotation=70)
plt.title("Valor por unidade da federacao")
plt.show()
```
